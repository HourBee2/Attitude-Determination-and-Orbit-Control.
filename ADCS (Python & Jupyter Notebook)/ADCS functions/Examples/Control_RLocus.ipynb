```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctl
```

```python
w_n = 1
damp = 1
```

```python
# Underdamped system
sys = ctl.tf([0, 0, 1], [0, 1, 0]) * \
    ctl.tf([0, 0, w_n ** 2], [1, 2*damp*w_n, w_n ** 2])
sys
```

```python
poles, zeros = ctl.pzmap(sys, plot=True, grid=True)
plt.show()
```

```python
ctl.rlocus(sys, xllim=[-5, 2], yllim=[-3, 3])
plt.show()
```

```python
T, yout = ctl.step_response(sys)  # , T=np.linspace(0, 5, 100))
plt.plot(T, yout)
plt.show()
```
